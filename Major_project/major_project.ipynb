```python
#5:Train and test variables
# we are going to seperate the data in train and test
# as data is too large we are going to split data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)  # splitting data---taking 20 % of data for testing
```

```python
#6:Normalisation/scaling - To be done only for inputs ---only applicable on multivariate datasets
```

```python
#7:Run a classifier , Regressor or clusterer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
model = LinearRegression()
```

```python
# 8:Fitting the model----(Mapping /plotting the inputs with the outputs)
model.fit(x_test,y_test) # plotting the values of x_test and y_test in the LinearRegression library
```

```python
# 9: Predict the Output
y_pred = np.round(model.predict(x_test)) # Using the input values, we predict the output
y_pred
```

```python
# Actual outputs
y_test = np.round(y_test)
y_test# As we can see difference is huge in actual and predicted output values
# Hence our Model is not linear
```

```python
accuracy_percent = accuracy_score(y_test,y_pred)*100
print(f"Accuracy:  {accuracy_percent}%")
```
